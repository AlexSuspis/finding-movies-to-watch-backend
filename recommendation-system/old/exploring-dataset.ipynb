```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv('input/small_dataset/movies.csv')
movies_df
movies_df.columns
```

```python
#movies_df.rename(columns={'id': 'movieId'}, inplace=True)
movies_df = movies_df[['movieId', 'title']]
movies_df
```

```python
ratings_df = pd.read_csv('input/small_dataset/ratings.csv')
ratings_df
```

```python
final_dataset = ratings_df.pivot(index='movieId', columns='userId', values='rating')
final_dataset
```

```python
#Data pre-processing
final_dataset.fillna(0, inplace=True)
final_dataset
```

```python
#Finding number of ratings each movie has
no_ratings_per_movie = ratings_df.groupby('movieId')['rating'].agg('count')
no_ratings_per_movie
```

```python
#finding number of movies each user rated
no_movies_voted_per_user = ratings_df.groupby('userId')['rating'].agg('count')
no_movies_voted_per_user
```

```python
#In order to filter out the noise, we are interested only in 

#1) users who have rated more than 50 movies
final_dataset = final_dataset.loc[:,no_movies_voted_per_user[no_movies_voted_per_user > 50].index]

#2) movies who have been rated by more than 10 users
final_dataset = final_dataset.loc[no_ratings_per_movie[no_ratings_per_movie > 10].index,:]

final_dataset
```
